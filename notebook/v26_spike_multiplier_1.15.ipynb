```python
import pandas as pd

print("--- APPLYING SPIKE MULTIPLIER TO v12 ---")

# 1. Load your absolute best submission
df = pd.read_csv('../submissions/submission_v12_23.747666.csv')

# 2. Identify the top 5% of extreme predictions (The Winter Spikes)
# (You can adjust the threshold, but 100 is usually a solid start for extreme grid prices)
threshold = 100.0  

# 3. Apply a 1.1x (10%) multiplier ONLY to the extreme spikes
spike_mask = df['target'] > threshold
df.loc[spike_mask, 'target'] = df.loc[spike_mask, 'target'] * 1.2

print(f"Boosted {spike_mask.sum()} extreme predictions by 10%.")

# 4. Save the post-processed submission
df.to_csv('../submissions/submission_v26.csv', index=False)
print("✅ Saved v23! Ready for submission.")
```
